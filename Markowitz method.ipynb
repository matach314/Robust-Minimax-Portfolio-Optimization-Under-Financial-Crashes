```python
import pandas as pd
import numpy as np
os.chdir(r"C:\Users\matra\Downloads")



# load
reliance = pd.read_csv("RELIANCE.NS_.csv")
tcs = pd.read_csv("TCS.NS.csv")
icici = pd.read_csv("ICICIBANK.NS.csv")
hdfc = pd.read_csv("HDFCBANK_NSE_1995-2025.csv")




# select correct columns
reliance = reliance[["Date","Adj Close"]]
tcs = tcs[["Date","Adj Close"]]
icici = icici[["Date","Adj Close"]]
hdfc = hdfc = hdfc[["Date","Close"]]


# rename columns
reliance.rename(columns={"Adj Close":"Reliance"}, inplace=True)
tcs.rename(columns={"Adj Close":"TCS"}, inplace=True)
icici.rename(columns={"Adj Close":"ICICI"}, inplace=True)
hdfc.rename(columns={"Close":"HDFC"}, inplace=True)

# merge datasets
data = reliance.merge(tcs,on="Date")
data = data.merge(icici,on="Date")
data = data.merge(hdfc,on="Date")

# drop missing
data = data.dropna()

# compute returns
returns = np.log(data.iloc[:,1:] / data.iloc[:,1:].shift(1)).dropna()

# mean and covariance
mu = returns.mean()
Sigma = returns.cov()

# minimax parameters
delta = 0.1
lam = 1

I = np.eye(len(mu))

# compute weights
w = np.linalg.inv(lam*Sigma + delta*I).dot(mu)

# normalize
w = w/sum(w)

print(w)
```

```python
portfolio_return = np.sum(w * m)
print(portfolio_return)
```

```python
portfolio_risk = np.sqrt(np.dot(w.T, np.dot(Sigma, w)))
print(portfolio_risk)
```

```python
sharpe_ratio = portfolio_return / portfolio_risk
print(sharpe_ratio)

```
